```python
# filling in / adding students details
students = []
others = "yes"

while True and others == "yes":
    name = input("Enter your name: ")
    age = input("Enter your age: ")
    while not age.isnumeric():
      age = input("Enter a valid age: ")
    gender = input("Enter your gender: ")
    while gender != "male" and gender != "female" and gender != "other":
      gender = input("Enter a valid gender: ")
    marks = input("Enter your marks: ")
    while not marks.isnumeric():
      marks = input("Enter a valid mark: ")

    student = {

      "name": name,
      "age": age,
      "gender": gender,
      "marks": marks

  }

    students.append(student)

    others = input("Add another student?(yes/no): ")
    print("\n")

    if others == "no":
      break

# viewing the student details

student_number = 1
print("=====Student Details=====")
for student in students:
  print("Student", student_number)
  print("Name:", student["name"])
  print("Age:", student["age"])
  print("Gender:", student["gender"])
  print("Marks:", student["marks"])
  print("\n")
  student_number = student_number + 1

# recording each students marks
print("=====Student marks=====")
for student in students:
  print(student["name"], "-" ,student["marks"])
print("\n")

# viewing failing scores
print("=====Failing scores=====")
fail = False

for student in students:
  if int(student["marks"])<50:
    print(student["name"], "-", student["marks"])
    fail = True

if fail == False:
  print("No failing scores")
  print("\n")


# exiting the program
if others == "no":
  print("Exiting program...")
```

```python
students = {}


while True:
    name = input("Enter student name (or press Enter to finish): ")

    if name == "":
        break

    students[name] = []


for name in students:
    mark = int(input(f"Enter mark for {name}: "))
    students[name].append(mark)


print("\nComplete Student Records")
for name in students:
    print(name, ":", students[name])


print("\nFailing Scores")
for name in students:
    failing = [mark for mark in students[name] if mark < 50]
    print(name, ":", failing)
```
