# .**loc** and .**iloc**


```python
df.loc[2, "age"] # row 2, column named age
# allows you to input the specific column name
```

```python
df.iloc[2,0] # third row, first column (age) (incorporates indexes)
# you input the index of the desired column
```

```python
df.iloc[5,0] # 6th row, first column(age)
```

```python
df.iloc[2,2]
```

```python
df.loc[0] # shows you info in the first row only
```

```python
df.loc[0:2] # shows you the first 3 rows
```

```python
df.iloc[0]
```

```python
df.iloc[0:2] # showa you the first and second row.....indices are applied
```
