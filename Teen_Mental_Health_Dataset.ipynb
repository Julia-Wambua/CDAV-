# Two conditions at once
*   and   **(&)**
*   or    **(|)**



```python
# female and 14
df[(df["age"]== 14) & (df["gender"] == "female")] # and(&)
# both conditions have to be true for you to get an output
```

```python
# female or 14
df[(df["age"] == 14) | (df["gender"] == "female")] # or
# either of the conditions can be true to get an output
```

# Not, and one-of-many
*   **(-)** means *NOT*
*   .**isin()** - *one of many*



```python
# Not age (-) means not
df[-(df["age"] == 14 )]
```

```python
# one of several, without a long chain of or's
df[df["age"].isin([14,15,16])] # gives you colums with age as 14,15 or 16
```

```python
df[-df["age"].isin([14,15,16])]
# gives rows of all ages EXCEPT 14,15 and 16
# (-) mean NOT
```

# Ranges and a readable alternative
*   **.between()** - inclusive at both ends
*  **.query()** - the same filter, written as a sentence



```python
# between - inclusive at both ends, NO EXCLUSIVE VALUES
df[df["age"].between(14,16)] # gives rows whose ages are 14,15 and 16
```

```python
# query - the same filter, written as a entence
df.query("14 <= age <= 16")
```
