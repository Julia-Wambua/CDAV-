```python
def authenticator(password):
  defined_pass = "secret"
  if password == defined_pass:
    return True
  else:
    return False
```

```python
passw = input("Enter  password: ")
if authenticator(passw):
  print("Access Granted")
else:
  print("Access Denied")
```

```python
def sum_diff(x,y):
  s = x+y
  d = x-y
  # returning multiple values from a function using a tuple
  return (s,d)

sum_of_both, diff_of_both = sum_diff(100,50)
print(sum_of_both)
print(diff_of_both)
```

```python
# this syntax is used to create a throw-away variable
_, diff_2 = sum_diff(100,50)
print(diff_2)

# that is why only 50 is printed since 100 has been disposed of.
```

 - Write a program that asks the user to enter his or her month and day of birth. Then

 - Write a program that reads a year from the user and displays the animal associated
with that year. Your program should work correctly for any year greater than or equal
to zero, not just the ones listed in the table.

```python
year = int(input("Enter a year: "))
zodiac = (year) % 12
if zodiac == 8:
  print("Your animal is a dragon")
elif zodiac == 9:
  print("Your animal is a snake")
elif zodiac == 10:
  print("Your animal is a horse")
elif zodiac == 11:
  print("Your animal is a sheep")
elif zodiac == 0:
  print("Your animal is a monkey")
elif zodiac == 1:
  print("Your animal is a rooster")
elif zodiac == 2:
  print("Your animal is a dog")
elif zodiac == 3:
  print("Your animal is a pig")
elif zodiac == 4:
  print("Your animal is a rat")
elif zodiac == 5:
  print("Your animal is a ox")
elif zodiac == 6:
  print("Your animal is a tiger")
elif zodiac == 7:
  print("Your animal is a hare")
else:
  print("Invalid input")
```
