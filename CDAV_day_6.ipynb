```python
# Python Workbook number 27
units = input("Choose a unit system(M for metric, I for imperial): ").upper()
# to determine the units to be used for the input
# and calculations


if units == "M":
  height = float(input("Enter your height(m): "))
  weight = float(input("Enter your weight(kg): "))
  BMI = weight / (height * height)
  print("Your BMI is: ", BMI)

elif units == "I":
  height = float(input("Enter your height(in): "))
  weight = float(input("Enter your weight(lbs): "))
  BMI = (weight  / (height * height))*703
  print("Your BMI is: ", BMI)

else:
  print("Invalid unit system")

```

# Methods and functions

```python
cols = (data[0:19].split(","))
print(cols)
# .append() allows you to add items to the end of the list
cols.append("age")
print(cols)

# .pop() removes an item from the end of the list
# gives the exact item it has removed
cols.pop()
print(cols)
cols.pop(1)
print(cols)

# .insert() allows you to insert an item at a given index
cols.insert(1,"gender")
print(cols)

# .remove() removes an item that is already exsisting in the list
# if the item is not present, you get an error
if "gender" in cols:# makes sure gender is in the list
  cols.remove("gender")
print(cols)

# .extend() allows you to add more items to the end of your list
# joins two or more lists
cols.extend(["age", "gender"])
print(cols)


# you can also use addition to extend your lists
cols = cols+["address", "email"]
print(cols)


# .clear() removes all items
# .index() gives the index of the first occurrence of an item
# .count() gives the number of occurrences of an item
cols.index("age")
cols = cols+["address", "email"]
print(cols)
print(cols.count("address"))
cols.index("address")

# .sort(allows arrangement from smallest to largest)
marks = [20,34,18,45,4,7]
#marks.sort(reverse=True) - standard method of reversing
marks.sort()
marks.reverse()
print(marks)
```

# Create a program that keeps asking the user for numbers until they press =. The program should then output:
*   highest value
*   lowest value
*   sum
*   average





```python
# APPLICATION OF LISTS, WHILE LOOPS AND IF STATEMENTS

numbers = []# the list is empty at the beginning
value = ""# at first no value has been entered
while value != "=":
  value = (input("Enter a number: "))
  print("You entered: ", value)
  if value != "=" and value.isnumeric():# checks whether the value entered
  # numeric. If not, it continues to prompt the user to enter values
  # if letters are entered, they will not be included in the calculations
    value = int(value)
    numbers.append(value)
  if value == "=":
    break

if value == "=":
  print("Numbers: ",numbers)
  print("Sum: ",sum(numbers))
  print("Maximum value: ",max(numbers))
  print("Minimum value: ",min(numbers))
  print("Average: ",sum(numbers)/len(numbers))











```

# Sets

```python
fruits = ["apple", "mango", "apple", "mango", "banana"]
print(fruits)
# get rids of repeated values
fruits2 = set(fruits)
print(fruits2)
```

```python
fruits2 = set(fruits)
{fruit.upper() for fruit in fruits2}
# order is sets is not maintained hence is UNORDERED
fruits2.add("kiwi")
print(fruits2)
# when using remove, the item has to exist in the set
fruits2.remove("apple")
print(fruits2)
# more foolproof then remove
fruits2.discard("kiwi")
print(fruits2)
```
