```python
body_temp_in_degrees_celcius = input("What is your body temperature? ")
body_temp = float(body_temp_in_degrees_celcius)
# we use float because temp varies and hence has decimal points,
# and because you cannot multiply a string and a float
body_temp_in_degrees_farenheit = ((body_temp * 1.8)+ 32)
print("Your body temperature in degrees celcius is", body_temp_in_degrees_celcius)
print("Your body temperature in degrees farenheit is", body_temp_in_degrees_farenheit)
```

```python
# it is better txo use floats since they are more accurate than integers
# and because one of the numbeers chosen may be in decimal form
number_1 = float(input("First Number: "))
number_2 = float(input("Second Number: "))
print("The sum of your numbers is", number_1 + number_2)
print("The difference of your numbers is", number_1 - number_2)
print("The product of your numbers is", number_1 * number_2)
print("The quotient of your numbers is", number_1 / number_2)
print("The remainder of your number is", number_1 % number_2)
print("The floor division of your number is", number_1 // number_2)
print("The power of your numbers is", number_1 ** number_2)
```
