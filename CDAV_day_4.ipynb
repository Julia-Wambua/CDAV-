```python
#36,39 and 45 (python workbook)
# Number 36
letter = input("Enter a letter of the alphabet: ")
if letter == "a" or letter == "e" or letter == "i" or letter =="o" or letter =="u":
  print("The letter entered is a vowel")
elif letter == ("y"):
  print("The letter entered is sometimes a vowel and sometimes a consonant")
else:
  print("The letter entered is a consonant")
```

```python
# Number 39
sound_level = int(input("Enter the sound level in decibels: "))
if sound_level > 130 :
  print("The sound level entered is louder than a Jackhammer.")
elif sound_level == 130:
  print("THis corresponds to a Jackhammer.")
elif sound_level > 106 and sound_level < 130:
  print("This is found between a Jackhammer and a Gas Lawnmower")
elif sound_level == 106:
  print("This correesponds to a Gas Lawnmower")
elif sound_level > 70 and sound_level < 106:
  print("This is found between an Alarm Clock and a Gas Lawnmower")
elif sound_level == 70:
  print("This corresponds to an Alarm Clock")
elif sound_level > 40 and sound_level < 70:
  print("This is found between a Quiet Room and an Alarm Clock")
elif sound_level == 40:
  print("This corresponds to a Quiet Room")
elif sound_level < 40:
  print("The sound level entered is quieter than a Quiet Room")
```
