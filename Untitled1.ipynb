```python
students = {}


while True:
    name = input("Enter student name (or press Enter to finish): ")

    if name == "":
        break

    students[name] = []


for name in students:
    mark = int(input(f"Enter mark for {name}: "))
    students[name].append(mark)


print("\nComplete Student Records")
for name in students:
    print(name, ":", students[name])


print("\nFailing Scores")
for name in students:
    failing = [mark for mark in students[name] if mark < 50]
    print(name, ":", failing)
```
