# Decision making using python
If and boolean statements

```python
is_raining = True
if is_raining :
  print("Stay indoors")
```

The statement is only executed if the is_raining variable is true.

Use not rather than the exclamation sign at times to avoid syntax errors.

```python
is_raining = True
if not is_raining :
  print("Go outside")
```

# Grade converter
It takes the grade and tells the rangue of marks

```python

grade = input("Enter the attained grade: ")
if grade == "A":
  print("You scored above 80")
elif grade == "B":
  print("You scored between 71 and 80")
elif grade == "C":
  print("You scored between 61 and 70")
elif grade == "D":
  print("You scored between 51 and 60")
elif grade == "E":
  print("You scored between 41 and 50")
elif grade == "F":
  print("You scored below 40")
else:
  print("Invalid grade")
```

# Write a program that takes a value from the user. The program should print
# Application of if elif else
*   divisible by 2 if the value is divisible by 2
*   divisible by 3 if the value is divisible by 3
*   divisible by both 2 and 3 if the value is divisible by both 2 and 3
*   otherwise, it should print not divisible by 2 and 3





```python
value = int(input("Enter a value: "))
if value % 2 == 0 and value % 3 == 0:
  print("Divisible by both 2 and 3")
# % indicates that the remainder you get when you divide the given value
# by either 2 or 3 is 0
elif value % 2 == 0:
  print("Divisible by 2")
elif  value % 3 == 0:
  print("Divisible by  3")
else:
  print("Not divisible by 2 and 3")

```
