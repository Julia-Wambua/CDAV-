# Application of if elif else
*   divisible by 2 if the value is divisible by 2
*   divisible by 3 if the value is divisible by 3
*   divisible by both 2 and 3 if the value is divisible by both 2 and 3
*   otherwise, it should print not divisible by 2 and 3





```python
value = int(input("Enter a value: "))
if value % 2 == 0 and value % 3 == 0:
  print("Divisible by both 2 and 3")
# % indicates that the remainder you get when you divide the given value
# by either 2 or 3 is 0
elif value % 2 == 0:
  print("Divisible by 2")
elif  value % 3 == 0:
  print("Divisible by  3")
else:
  print("Not divisible by 2 and 3")

```

# Slicing
*   The first number in the square brackets indicates where to start
*   In regard to the second number, slicing ends just before the second number indicated
*   If you don't specify where to start or stop, it starts from the beginning and ends at the end of the word






```python
text = "python"
print(text[0:2])
print(text[2:])
print(text[:4])
```
